```python
#Importing packages
import pandas as pd
import pickle
import nltk
nltk.download('punkt')
```

```python
# Opening the pickle containing beer data
with open('biertjes_scraped_1150_v1.pkl', 'rb') as f:
    beerdata = pickle.load(f)
beerdata
```

```python
# Loading flavor data
flavors = pd.read_csv('Flavor.csv')
```

```python
beerdesc = beerdata[['id','description']]
```

```python
# Tokenizing the beer discription and counting every that is in the flavorlist
beerlist = []
flavorlist = flavors['name'].tolist()

for index,row in beerdesc.iterrows():
    a = 0
    
    flavordict = {'id': row['id'],
               'flavors': []}
    
    for word in nltk.word_tokenize(row['description']):
        
        if(word in flavorlist):
            flavordict['flavors'].append(word)
                
        if(a == (len(nltk.word_tokenize(row['description'])) - 1)):
            beerlist.append(flavordict)
        a = a + 1
```

```python
for element in beerlist:
    element['flavors'] = list(set(element['flavors']))
```

```python
beerlist
```

```python
#Nested dictionary to list
idlist = []
flavorslist = []
for element in beerlist:
    if(len(element['flavors']) > 0):
        for flavor in element['flavors']:
            idlist.append(element['id'])
            flavorslist.append(flavor)
            
df = pd.DataFrame({'id':idlist,
                  'flavor': flavorslist})
```

```python
df
```

```python
final_df = pd.merge(beerdata,df,on='id',how='outer')
```

```python
final_df.to_csv('flavortagged_beers.csv')
```
